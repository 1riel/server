```python
# TODO: create bucket with policy

# create buckets
if not s3.bucket_exists(MINIO_BUCKET_PUBLIC):
    s3.make_bucket(MINIO_BUCKET_PUBLIC)

if not s3.bucket_exists(MINIO_BUCKET_PRIVATE):
    s3.make_bucket(MINIO_BUCKET_PRIVATE)


# set bucket policy
policy = f"""{{
    "Version": "2012-10-17",
    "Statement": [
        {{
            "Effect": "Allow",
            "Principal": {{"AWS": ["*"]}},
            "Action": ["s3:GetObject"],
            "Resource": ["arn:aws:s3:::{MINIO_BUCKET_PUBLIC}/*"]
        }},
        {{
            "Effect": "Allow",
            "Principal": {{"AWS": ["*"]}},
            "Action": ["s3:ListBucket"],
            "Resource": ["arn:aws:s3:::{MINIO_BUCKET_PUBLIC}"]
        }}
    ]
}}"""


s3.set_bucket_policy(MINIO_BUCKET_PUBLIC, policy)
```

```python
# list all files in the bucket
objects = s3.list_objects(MINIO_BUCKET_PUBLIC, recursive=True)
# pprint([obj.object_name for obj in objects])
for obj in objects:
    print(obj.object_name)
    if obj.object_name.startswith("assets/"):
        cvt.storage_to_thumbnail(obj.object_name, 100)
        cvt.storage_to_thumbnail(obj.object_name, 200)
```
